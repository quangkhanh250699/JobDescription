```python
import pandas as pd 
import findspark
findspark.init()

import _mssql
from pyspark import SparkContext, SparkConf, SQLContext
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName('Job_description') \
                            .master('local') \
                            .getOrCreate()
spark
```

```python
df = spark.read.option('multiLine', True) \
                .option('path', '../data/job_description.csv') \
                .option('header', True) \
                .option('quote', '"') \
                .format('csv') \
                .load() 

df.sample(fraction=0.1).show(n=2, truncate=False, vertical=True)
```

```python
a = [i[0] for i in df.select('job_level').collect()]
```

```python
with open('a.txt', 'w') as f: 
    for i in a: 
        f.write(str(i) + '\n') 
```

```python
with open('a.txt', 'r') as f: 
    job_level = f.readlines()
job_level = list(map(lambda x: x.replace('\n', ''), job_level))
set(job_level)
```
